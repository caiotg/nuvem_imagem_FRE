```python
from pypdf import PdfReader
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
reader = PdfReader("releaseGerdau1T2023.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
conectivos = set(STOPWORDS)
conectivos.update(["da", "meu", "em", "e", "você", "de", "ao", "os", 'na', 'o', '4T21', 'para', 'milhões',
                  'resultado', 'que', 'nas', 'dos', "n", "p", "4T22", "1T23", "1T22", "trimestre", "Brasil", "América"])
```

```python
nuvemPalavras = WordCloud(stopwords= conectivos, background_color= 'black', width= 1920, height= 1080).generate(text)
```

```python
fig, ax = plt.subplots()
ax.imshow(nuvemPalavras)
ax.set_axis_off()
```
